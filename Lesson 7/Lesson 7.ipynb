### Revisiting print

We already know how tu use print for:

- strings
- strings containing special characters
- integers
- concatenation of several strings or stirngs and integers
- lists

```python
print('Hello World!')
print('one\ntwo\nthree')
print(25*4)
print('Hello' + ' ' + 'Nazly')
print('If you multiply 25 and 4, you will get: ' + str(25*4))
```

### Is it possible to print not to a screen, but to a file?

- yes, and this is easy!


```python
f = open('test.txt', 'w')
f.write('Hello World!\n'*5)
f.close()   
```

### What about reading from a file?

two main options:
- as a whole text
- line by line

```python
f = open('test.txt', 'r')

file_content = f.read()
print(file_content)
```

```python
f = open('test.txt', 'r')

for line in f:
    print(line)
```

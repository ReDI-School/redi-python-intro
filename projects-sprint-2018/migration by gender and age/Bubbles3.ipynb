```python
import csv

def print_data_from_csv(filename):
    with open (filename) as f:
        data = csv.reader(f)
        count = 1
        for x in data:
            print (x)
            count += 1
            if count >10:
                break

                
print_data_from_csv("project.csv")
```

```python
#Getting the name of the countries in a dictionary
def load_names (file_name):
    with open (file_name) as f:
        data = csv.reader(f)
        
        mydata = {}
        for row in data:
            x = mydata.get(row[1],row[0])
            mydata[row[1]] = x
            
            
        return mydata

country_name = load_names("countries.codes.csv")
import pprint
pprint.pprint (country_name)
```

```python
#Total per group per country 
def load_data(file_name):
    with open (file_name) as f:
        data = csv.reader(f)
        next(data)
        my_list = []
        for x in data:
            my_list.append(x)
    return my_list    

    
def total_per_group (my_list):
        my_dic = {}
        for line in my_list:
            country = line[1]
            X = my_dic.get(country, [0,0,0,0,0])
            for i in range (5):
                X[i] += int(line[i+2])
            my_dic[country] = X
        return my_dic
    
my_data = load_data('project.csv')
import pprint
great_dictionary = total_per_group(my_data)
pprint.pprint (great_dictionary)
```

```python
from operator import itemgetter

greatest_list = sorted(great_dictionary.items(), key=itemgetter(1), reverse=True)

ten_items = greatest_list[:10]
for t in ten_items:
    print('Entry:', t[0])
    print('Entry:', t[1])
```

```python
#getting the total per group 
def load_data(file_name):
    with open (file_name) as f:
        data = csv.reader(f)
        next(data)
        my_list = []
        for x in data:
            my_list.append(x)
    return my_list    

    
def total_per_group (index, my_list):
        second_list = []
        for line in my_list:
            second_list.append(int(line[index]))
            final_list = sum(second_list)
        return final_list
    
my_data = load_data('project.csv')
great_list = total_per_group(3,my_data)
print ('Male:',great_list)
great_list = total_per_group(4,my_data)
print ('Female:',great_list)
great_list = total_per_group(5,my_data)
print ('AM:',great_list)
great_list = total_per_group(6,my_data)
print ('UAM:',great_list)
```

```python
import folium

from  geopy.geocoders import Nominatim

europe = [48.5260, 21.2551]
center = [26.851854, 16.302643]

fmap = folium.Map(location=center, zoom_start=4)


def plot_genders(fmap, data, range):
    gloc = Nominatim()
    divider = 2000
    for x in data[:range]:
        code = x[0]
        if code == 'OOO':
            continue
        name = country_name[code]
        total_females = x[1][2]/divider
        total_males = x[1][1]/divider
        total_AM = x[1][3]/divider
        total_UAM = x[1][4]/divider
        #print('Country code', code, 'and name', name,
        #      'total males', total_males, 'and females', total_females)
        country_location = gloc.geocode(name)
        #print(country_location.latitude, country_location.longitude)
        folium.CircleMarker(location=[country_location.latitude, country_location.longitude],
                            popup=str(total_females*divider),
                            radius=total_females, color='red', fill=False).add_to(fmap)
        folium.CircleMarker(location=[country_location.latitude, country_location.longitude],
                            popup=str(total_males*divider),
                            radius=total_males, color='blue', fill=False).add_to(fmap)
        folium.CircleMarker(location=[country_location.latitude, country_location.longitude],
                            popup=str(total_AM*divider),
                            radius=total_AM, color='green', fill=False).add_to(fmap)
        folium.CircleMarker(location=[country_location.latitude, country_location.longitude],
                            popup=str(total_UAM*divider),
                            radius=total_UAM, color='yellow', fill=False).add_to(fmap)
        
plot_genders(fmap, greatest_list, 20)
fmap
```
